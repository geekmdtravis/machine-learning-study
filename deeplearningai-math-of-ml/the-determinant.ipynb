# The Determinant, and it's Application in 3x3 Matrices

Recall that the determinate of a 2x2 matrix is defined as:

$$\begin{vmatrix} a & b \\ c & d \end{vmatrix} = ad - bc$$

In a 3x3 matrix, the determinant is not so simply defined. Instead, we must use the following formula:

$$\begin{vmatrix} a & b & c \\ d & e & f \\ g & h & i \end{vmatrix} = a\begin{vmatrix} e & f \\ h & i \end{vmatrix} - b\begin{vmatrix} d & f \\ g & i \end{vmatrix} + c\begin{vmatrix} d & e \\ g & h \end{vmatrix}$$

While this seems confusing, we're actually just following a pattern of identifying diagonals and anti-diagonals that can be summed.

For the following 3x3:

$$
\begin{gather}
\begin{vmatrix} 
a & b & c \\ 
d & e & f \\ 
g & h & i 
\end{vmatrix}
\end{gather}
$$

The diagonals are $\begin{vmatrix} a & e & i \end{vmatrix}$, $\begin{vmatrix} b & f & g \end{vmatrix}$, and $\begin{vmatrix} c & d & h \end{vmatrix}$.

And the anti-diagonals are $\begin{vmatrix} c & e & g \end{vmatrix}$, $\begin{vmatrix} a & f & h \end{vmatrix}$, and $\begin{vmatrix} b & d & i \end{vmatrix}$.

The diagnoals start from the top left, and the anti-diagonals start from the top right. you simply translate and wrap around the matrix.
You then multiply the diagonals and anti-diagonals by the sign $(-1)^{i+j}$, where $i$ is the row and $j$ is the column. This is the same as the sign of the cofactor. Also note that functionally, the sign simply alternates as you move down the matrix since $i$ and $j$ are always increasing and will ever alternate between even and odd, which when applied as an exponent to a negative value, will alternate between positive and negative.

## Example: Question 1
Find the determinant of the following matrix:

$$\begin{vmatrix} 
1 & 0 & 1 \\ 
0 & 1 & 0 \\ 
3 & 3 & 3 
\end{vmatrix}$$

### Solution (long version with no shortcuts)

$$\begin{gather}
(1 * 1 * 3) + (0 * 0 * 3) + (1 * 0 * 3)  = 3 | \text{Diagonals}\\
(1 * 1 * 3) + (0 * 0 * 3) + (1 * 0 * 3)  = 3 | \text{Anti-diagonals}\\
\end{gather}$$

Subtracting the anti-diagonals from the diagonals, we get $3 - 3 = 0$, so this is a _linearly dependent_ and _singular_ system of equations.

##  The Upper Triagonal Matrix Shortcut
In the event that you have a matrix that is upper triagonal, you can simply multiply the diagonals together to get the determinant. This is because the anti-diagonals will always be zero, and thus the sum of the diagonals will be the same as the sum of the diagonals minus the sum of the anti-diagonals.

Example:

$$\begin{vmatrix}
1 & 2 & 3 \\
0 & 4 & 5 \\
0 & 0 & 6
\end{vmatrix} = 1 * 4 * 6 = 24$$

If the matrix a upper triagonal AND has a zero in the diagonal, then the determinant is zero. This is because the diagonals will always be zero, and thus the sum of the diagonals will be the same as the sum of the diagonals minus the sum of the anti-diagonals.

Example:

$$\begin{vmatrix}
0 & 2 & 3 \\
0 & 4 & 5 \\
0 & 0 & 6
\end{vmatrix} = 0 * 4 * 6 = 0$$

## Row Eshelon Form Shortcut

If a matrix is in row eshelon form, then the determinant is the product of the diagonal elements. This is because the anti-diagonals will always be zero, and thus the sum of the diagonals will be the same as the sum of the diagonals minus the sum of the anti-diagonals. So, attempting to manipulate the matrix into row eshelon form is a good strategy for finding the determinant.

Here is an example of Row Eshelon Form:

$$\begin{vmatrix}
1 & 2 & 3 \\
0 & 4 & 5 \\
0 & 0 & 6
\end{vmatrix} = 1 * 4 * 6 = 24$$

Here is a matrix that is not  yet in row eshelon form:

$$\begin{vmatrix}
1 & 2 & 3 \\
0 & 4 & 5 \\
0 & 1 & 6
\end{vmatrix}$$

We can manipulate it into row eshelon form by multiplying the third row by -4 and adding it to the second row:

$$\begin{vmatrix}
1 & 2 & 3 \\
0 & 4 & 5 \\
0 & 1 & 6
\end{vmatrix} = \begin{vmatrix}
1 & 2 & 3 \\
0 & 4 & 5 \\
0 & 0 & -19
\end{vmatrix}$$

The determinant is now the product of the diagonal elements, which is $1 * 4 * -19 = -76$.
